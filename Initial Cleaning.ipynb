```python
all = pd.read_excel("C:/Uni - Year 4/ACTL4001/All Data.xlsx", sheet_name=None)
clean = {}
for sheet_name, df in all.items():

    # Fix numeric columns
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            df[col] = df[col].abs()

            # Apply global upper bound
            if col in cap and isinstance(cap[col], (int, float)):
                df[col] = df[col].clip(upper=cap[col])

            # Apply global lower bound
            if col in low and isinstance(low[col], (int, float)):
                df[col] = df[col].clip(lower=low[col])

    # Apply sheet-level bounds (upper + lower)
    if sheet_name in cap and isinstance(cap[sheet_name], dict):
        for c, limit in cap[sheet_name].items():
            if c in df.columns:
                df[c] = df[c].clip(upper=limit)

    if sheet_name in low and isinstance(low[sheet_name], dict):
        for c, limit in low[sheet_name].items():
            if c in df.columns:
                df[c] = df[c].clip(lower=limit)

    df = df.dropna()
    clean[sheet_name] = df
clean
```
